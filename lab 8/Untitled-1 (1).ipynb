```python
import math

def minmax(depth, index, is_maximizing, scores, target_depth):

    if depth == target_depth:
        return scores[index]

    if is_maximizing:
        left_value = minmax(depth + 1, index * 2, False, scores, target_depth)
        right_value = minmax(depth + 1, index * 2 + 1, False, scores, target_depth)
        return max(left_value, right_value)

    else:
        left_value = minmax(depth + 1, index * 2, True, scores, target_depth)
        right_value = minmax(depth + 1, index * 2 + 1, True, scores, target_depth)
        return min(left_value, right_value)


scores = [3, 5, 2, 9, 3, 5, 2, 9]

tree_depth = int(math.log(len(scores), 2))

print("Optimal Value:", minmax(0, 0, True, scores, tree_depth))
```
