```python
import os

# File to store tasks
FILE_NAME = "tasks.txt"

# Load existing tasks
def load_tasks():
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r") as f:
            return [line.strip() for line in f.readlines()]
    return []

# Save tasks
def save_tasks(tasks):
    with open(FILE_NAME, "w") as f:
        for task in tasks:
            f.write(task + "\n")

tasks = load_tasks()

print("\n --- Welcome to Your Personal To-Do List --- ")
print("You can: add | view | remove | done | exit")
print("Type your choice and hit Enter!\n")

while True:
    choice = input(" What would you like to do? ").strip().lower()

    if choice == "add":
        task = input(" Write your task: ").strip()
        if task:
            tasks.append(task)  # Task starts as not done
            save_tasks(tasks)
            print(f" Great! '{task}' has been added to your list.")
        else:
            print("⚠ Oops! You didn't type anything.")

    elif choice == "view":
        if tasks:
            print("\n Here are your tasks:")
            for i, task in enumerate(tasks, start=1):
                print(f"{i}. {task}")
        else:
            print("\n Your list is empty. Add something to get started!")

    elif choice == "remove":
        if tasks:
            print("\n Here are your tasks. Which one do you want to remove?")
            for i, task in enumerate(tasks, start=1):
                print(f"{i}. {task}")
            try:
                num = int(input("Enter the task number to remove: "))
                if 1 <= num <= len(tasks):
                    removed = tasks.pop(num-1)
                    save_tasks(tasks)
                    print(f" '{removed}' has been removed from your list.")
                else:
                    print(" That number is not on the list.")
            except ValueError:
                print("Please enter a valid number.")
        else:
            print(" Nothing to remove. Your list is empty.")

    elif choice == "done":
        if tasks:
            print("\n Which task do you want to mark as done?")
            for i, task in enumerate(tasks, start=1):
                print(f"{i}. {task}")
            try:
                num = int(input("Enter the task number: "))
                if 1 <= num <= len(tasks):
                    if "[DONE]" in tasks[num-1]:
                        print("This task is already marked as done.")
                    else:
                        tasks[num-1] = tasks[num-1] + " [DONE]"
                        save_tasks(tasks)
                        print(f" Task '{tasks[num-1]}' marked as done!")
                else:
                    print(" That number is not on the list.")
            except ValueError:
                print(" Please enter a valid number.")
        else:
            print(" No tasks to mark as done.")

    elif choice == "exit":
        print("\n Goodbye! Your tasks are saved. Here's your final list:")
        for i, task in enumerate(tasks, start=1):
            print(f"{i}. {task}")
        break

    else:
        print(" I didn’t understand that. Please type: add | view | remove | done | exit")
```
