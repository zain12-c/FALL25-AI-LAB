## Split Data into Training & Testing Sets

```python

df = pd.read_excel("nutritrack_dataset_600.xlsx")
df = df.dropna()

X = df[["meal_description", "cooking_method"]]
y = df["estimated_calories"].astype(float)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.10, random_state=42
)
```

## Data Preprocessing

```python
preprocess = ColumnTransformer(
    transformers=[
        ("text", TfidfVectorizer(max_features=3000, ngram_range=(1,1)), "meal_description"),
        ("cook", OneHotEncoder(handle_unknown="ignore"), ["cooking_method"])
    ],
    remainder="drop"
)
```

## Define Machine Learning Model

```python
model = RandomForestRegressor(
    n_estimators=500,
    max_depth=40,
    min_samples_split=4,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)
```

## Full Pipeline

```python
pipeline = Pipeline([
    ("preprocess", preprocess),
    ("model", model)
])
```

## Train the Model

```python
pipeline.fit(X_train, y_train)
```

## Make Predictions

```python
preds = pipeline.predict(X_test)
```

## Model Performance



```python
print("\n--- MODEL PERFORMANCE ---")
print("MAE :", mean_absolute_error(y_test, preds))
print("MSE :", mean_squared_error(y_test, preds))
print("R²  :", r2_score(y_test, preds))
```

## Save Model

```python
joblib.dump(pipeline, "model.joblib")
print("Model saved as model.joblib")
```
