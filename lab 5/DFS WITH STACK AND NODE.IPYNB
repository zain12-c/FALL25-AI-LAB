```python
# Node class
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

# Depth-Limited Search
def dls(node, target, limit):
    if node is None:
        return False

    if node.data == target:
        print(f"Found: {node.data}")
        return True

    if limit <= 0:
        return False

    # Search left and right children with reduced limit
    return (dls(node.left, target, limit - 1) or
            dls(node.right, target, limit - 1))

# Example usage
if __name__ == "__main__":
    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)

    target = 5
    depth_limit = 2

    print(f"DLS (searching {target} with limit {depth_limit}):")
    found = dls(root, target, depth_limit)
    if not found:
        print("Not found within depth limit")
```
